```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
d = pd.read_csv("canada_per_capita_income.csv") 
d.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(d.year,d.per_capita_income,color='red',marker='+')
```

```python
new_df = d.drop('per_capita_income',axis='columns')
new_df.head()
```

```python
per_capita_income = d.per_capita_income
per_capita_income.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
```

```python
reg.predict([[2020]])
```
